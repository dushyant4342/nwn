```python
data1_1 = data[data['Election Year'] == '2005-Feb']
data1_2 = data[data['Election Year'] == '2005-Oct']
data1_3 = data[data['Election Year'] == '2010']
data1_4 = data[data['Election Year'] == '2015']

data1__1 = data1_1[data1_1['Position'] == 1]
data1__2 = data1_2[data1_2['Position'] == 1]
data1__3 = data1_3[data1_3['Position'] == 1]
data1__4 = data1_4[data1_4['Position'] == 1]

data1_1
```

```python
data1_1 = data1_1.groupby('Party').sum()
```

```python
data1_1.head()
```

```python
data1_1=data1_1[data1_1['Votes']==data1_1['Votes'].max()]
data1_1=data1_1.reset_index()
```

```python
data1_1.head()
```

```python
parties1 = list(data1__1['Party'].values)
winner1 = data1_1['Party'][0]
print(winner1)
```

```python
seats1 = list(data1__1[data1__1['Party'] == winner1]['AC name'])
seats1
```

```python
data1_2 = data1_2.groupby('Party').sum()
data1_2=data1_2[data1_2['Votes']==data1_2['Votes'].max()]
data1_2=data1_2.reset_index()
```

```python
parties2 = list(data1__2['Party'].values)
winner2 = data1_2['Party'][0]
seats2 = list(data1__2[data1__2['Party'] == winner2]['AC name'])

data1_3 = data1_3.groupby('Party').sum()
data1_3=data1_3[data1_3['Votes']==data1_3['Votes'].max()]
data1_3=data1_3.reset_index()

parties3 = list(data1__3['Party'].values)
winner3 = data1_3['Party'][0]
seats3 = list(data1__3[data1__3['Party'] == winner3]['AC name'])

data1_4 = data1_4.groupby('Party').sum()
data1_4=data1_4[data1_4['Votes']==data1_4['Votes'].max()]
data1_4=data1_4.reset_index()


parties4 = list(data1__4['Party'].values)
winner4 = data1_4['Party'][0]
seats4 = list(data1__4[data1__4['Party'] == winner4]['AC name'])
```

```python
# Common seats in all 4 elections
from functools import reduce
res = list(reduce(lambda i, j: i & j, (set(x) for x in [seats1, seats2, seats3, seats4]))) 
print(res)
```
