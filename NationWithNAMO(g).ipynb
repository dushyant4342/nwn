```python
data1 = data.copy()

for i in range(12301):
    row = data[i:i+1]
    data1.at[i:i+1, ['Votes %']] = float((row['Votes %'].values[0])[:-1])
    
data1 = data1[data1['Position'] <= 2]
data1
```

```python
data1.reset_index
data1

res = []
for i in range(len(data1.index)-1):
    idx = data1.index[i]
    if abs(data1['Votes %'][idx] - data1['Votes %'][data1.index[i+1]]) < 10:
        res.append([data1['Election Year'][idx], data1['AC name'][idx]])
    i += 1
           
print(res[:-1])
```
