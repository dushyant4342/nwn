```python
data1 = data.copy()

for i in range(12301):
    row = data[i:i+1]
    data1.at[i:i+1, ['Votes %']] = float((row['Votes %'].values[0])[:-1])
    
data1
```

```python
data1.groupby('Election Year').sum()
```

```python
data1 = data1[data1['Votes %'] > 10] 
data1_1 = data1[data1['Election Year'] == '2005-Feb'].groupby('AC name').count()['Party']
data1_1.head()
```

```python
data1_2 = data1[data1['Election Year'] == '2005-Oct'].groupby('AC name').count()['Party']
data1_2.head()
```

```python
data1 = data1[data1['Votes %'] > 10] 
data1_1 = data1[data1['Election Year'] == '2005-Feb'].groupby('AC name').count()['Party']
data1_2 = data1[data1['Election Year'] == '2005-Oct'].groupby('AC name').count()['Party']
data1_3 = data1[data1['Election Year'] == '2010'].groupby('AC name').count()['Party']
data1_4 = data1[data1['Election Year'] == '2015'].groupby('AC name').count()['Party']

data1_1
res = []
for i in range(min(data1_1.count(), data1_2.count(), data1_3.count(), data1_4.count())):
    flag = 1
    if data1_1[data1_1.index[i]] < 3:
        flag += 1
    if data1_2[data1_2.index[i]] < 3:
        flag += 1
    if data1_3[data1_3.index[i]] < 3:
        flag += 1
    if data1_4[data1_4.index[i]] < 3:
        flag += 1
    
    if flag == 1:
        res.append(data1_1.index[i])
        
print(res)
```
