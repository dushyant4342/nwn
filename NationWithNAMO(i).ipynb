```python
data1 = data[data['Election Year'] == '2005-Feb']
data1 = data1[data1['Position'] == 1]

data1
```

```python
from collections import Counter
parties = data1['Party'].values
dict1 = Counter(parties)
dict1 = {k: v for k, v in sorted(dict1.items(), key=lambda item: item[1])}
del dict1['Independent']
print(dict1)
parties = list(dict1.keys())[-8:]
print('To Track:')
print(parties)
```

```python
names = list(set(data['AC name'].values))
print(names)
```

```python
res = []
for party in parties:
    for name in names:
        data_temp = data[data['Party'] == party]
        data_temp = data_temp[data_temp['AC name'] == name]
        if len(data_temp.index) == 4:
            if data_temp['Votes'][data_temp.index[0]] < data_temp['Votes'][data_temp.index[1]]:
                if data_temp['Votes'][data_temp.index[1]] < data_temp['Votes'][data_temp.index[2]]:
                    if data_temp['Votes'][data_temp.index[2]] < data_temp['Votes'][data_temp.index[3]]:
                        res.append([party, name])
                        
print(res)
```
