```python
import pandas as pd
```

```python
filename = 'NationWithNAMO.csv'
data = pd.read_csv(filename)
```

```python
data.head()
```

```python
data1 = data.groupby(['Election Year','Party']).sum()
```

```python
data1.head()
```

```python
data1 = data1.reset_index()
data1.head()
```

```python
data1 = data1.drop(['Position','AC No'],axis=1)
data1.head()
```

```python
result = data1.groupby('Election Year', group_keys=False).apply(lambda x: x.nlargest(8, 'Votes'))
```

```python
result = result.reset_index(drop=True)
```
